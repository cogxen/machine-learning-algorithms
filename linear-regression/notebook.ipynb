# Linear Regression Algorithm

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

## Sample Data

```python
# Create a sample data
num_cupcakes = [10, 20, 30, 40, 50]
money_made = [25, 70, 38, 110, 65]
```

Let's have a look at the relationship between the number of cupcakes sold and the money made.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=num_cupcakes, y=money_made)
plt.title('Number of Cupcakes Sold vs Money Made')
plt.xlabel('Number of Cupcakes Sold')
plt.ylabel('Money Made')
plt.show()
```

## Linear Regression Model

Now let's train a machine learning model to predict the money made based on the number of cupcakes sold.

```python
# Reshape the data
x = np.array(num_cupcakes).reshape(-1, 1)
y = np.array(money_made)

# Create a linear regression model
model = LinearRegression().fit(x, y)
```

### Model Evaluation

Let's evaluate the model by predicting the money made for the number of cupcakes sold.

```python
# Predict the money made
new_num_cupcakes = [[60]]
predicted_money_made = model.predict(new_num_cupcakes)

# Print the predicted money made
print(f'Predicted Money Made: {predicted_money_made[0]}')
```
