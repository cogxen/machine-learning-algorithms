# House Price Prediction

A **simple linear regression** model to predict house prices based on house sizes.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.linear_model import LinearRegression
```

## Data Preparation

```python
# House sizes (square feet)
house_sizes = [1000, 1500, 2000, 2500, 3000]

# House prices ($)
house_prices = [62728, 92272, 95105, 21852, 23221]
```

## Data Visualization

Let's have a look at the relationship between house sizes and prices and the linearity of the data.

```python
# Check the linearity of the data
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)
print(f'R-squared: {r_value**2:.2f}')

# Plot the regression line
plt.figure(figsize=(10, 5))
plt.scatter(house_sizes, house_prices, color='blue', label='Data')
plt.plot(house_sizes, slope*np.array(house_sizes) + intercept, color='red', label='Regression Line')
plt.xlabel('House Size (sqft)')
plt.ylabel('House Price ($)')
plt.title('House Size vs. Price')
plt.legend()
plt.show()
```

## Model Training

Let's train a simple linear regression model.

```python
# Convert lists to numpy arrays
x = np.array(house_sizes).reshape(-1, 1)
y = np.array(house_prices)

# Create a linear regression model
model = LinearRegression().fit(x, y)

# Predict house prices for the range of sizes
x_range = np.linspace(min(house_sizes), max(house_sizes), 100).reshape(-1, 1)
y_pred = model.predict(x_range)
```

## Model Evaluation

```python
# Predict house prices for the range of sizes
new_house_sizes = [[2000], [2500]]
predicted_prices = model.predict(new_house_sizes)

# Show the predicted prices
for size, price in zip(new_house_sizes, predicted_prices):
    print(f'Predicted price for a house of size {size[0]} sqft: ${price:.2f}')
```
